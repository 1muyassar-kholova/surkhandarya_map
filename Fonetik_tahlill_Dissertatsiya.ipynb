```python
import re

def analyze_dialect(text):
    # Define the rules for dialect analysis
    rules = {
        r'a': 'A, a – til orqa, quyi keng, lablangan variant. Misol: Xatna – xatnä.',
        r'ä': 'A, a ~ Ä, ä – til oldi, quyi keng, lablangan variant. Misol: taram-taram – täräm-täräm.',
        r'ā': 'A, a, (O, o) ga mos keladi, Ā, ā – til orqa, quyi keng, lablangan variant. Misol: āsmān, māma.',
        r'i': 'I, i ~ til oldi, qisqa, yuqori-tor, lablanmagan unli. Misol: chiroyli – čirājli.',
        r'ï': 'I, i ~ Ï, ï – til orqa, qisqa, yuqori-tor, lablanmagan unli. Misol: topishmog‘imiz – tāpišmāγïmïz.',
        r'ï:': 'I, i ~ Ï:, ï: – til orqa, cho‘ziq, yuqori-tor, lablanmagan unli. Misol: rozγār qï:p.',
        r'i:': 'I:, i: ~ til oldi, cho‘ziq, yuqori-tor, lablanmagan unli. Misol: bilqillaydi – bi:lqi:lläjdi.',
        r'u': 'U, u ~ U, u – til orqa, yuqori tor, lablangan variant unli. Misol: -urchiq – určïq.',
        r'ü': 'U, u ~ Ü, ü – til oldi, yuqori tor, lablangan variant unli. Misol: suyakka – süjäkä.',
        r'o‘': 'O‘, o‘ ~ O, o – til orqa, o‘rta keng, lablangan variant unli. Misol: qo‘y – qoj.',
        r'ö': 'O‘, o‘ ~ Ӧ, ӧ – til oldi, o‘rta keng, lablangan variant unli. Misol: toy – töj.',
        r'ŋ': 'ng ~ ŋ – sayoz til orqa, portlovchi, jarangli undosh. Misol: uyning – üjdiŋ.',
        r'ǯ': 'J, j ~ Ǯ, ǯ – til oldi, portlovchi, jarangli undosh. Misol: janda – Ǯändä.',
        r'č': 'Ch, ch ~ Č, č – til oldi, qorishiq, jarangsiz undosh. Misol: chirqillaydi – čärqilläjdi.',
        r'š': 'Sh, sh ~ Š, š – til oldi, sirg‘aluvchi, jarangsiz undosh. Misol: shunday – šündäj.',
        r'γ': 'G‘, g‘ ~ γ – chuqur til orqa, sirg‘aluvchi, jarangli undosh. Misol: qog‘oz – qaγaz, bo‘sag‘ada – bosaγadä.',
        r'j': 'J, j ~ Y, y – til oldi, sirg‘aluvchi, jarangli undosh. Misol: xushomadgo‘y – xušāmӓdgӧj.'
    }

    # Analyze the text based on the rules
    analysis = {}
    for pattern, description in rules.items():
        matches = re.findall(pattern, text, re.IGNORECASE)
        if matches:
            analysis[description] = matches

    return analysis

# Input text with examples
text = """
Xatna – xatnä, taram-taram – täräm-täräm, āsmān, māma, chiroyli – čirājli,
topishmog‘imiz – tāpišmāγïmïz, rozγār qï:p, bilqillaydi – bi:lqi:lläjdi,
-urchiq – určïq, suyakka – süjäkä, qo‘y – qoj, toy – töj, uyning – üjdiŋ,
janda – Ǯändä, chirqillaydi – čärqilläjdi, shunday – šündäj, qog‘oz – qaγaz,
bo‘sag‘ada – bosaγadä, xushomadgo‘y – xušāmӓdgӧj
"""

# Analyze the text
result = analyze_dialect(text)

# Print the analysis
for description, matches in result.items():
    print(f"{description}: {matches}")
```

```python
# O‘zbek tilidagi tovush o‘zgarishlari uchun lug‘at
sound_rules = {
    # Unlilar
    'a': ['a'],       # til orqa/quyi keng/lablangan
    'ä': ['ä'],       # til oldi/quyi keng/lablangan
    'ā': ['ā'],       # til orqa/qisman lablangan/quyi keng
    'i': ['i', 'ï', 'ï:'],  # til oldi va orqa/qisqa-cho‘ziq/lablanmagan
    'u': ['u', 'ü'],  # til orqa va oldi/yuqori tor/lablangan
    'o‘': ['o', 'ö'], # til orqa va oldi/o‘rta keng/lablangan

    # Undoshlar
    'ng': ['ŋ'],      # sayoz til orqa/portlovchi/jarangli
    'j': ['y'],       # til oldi/sirg‘aluvchi/jarangli
    'ch': ['č'],      # til oldi/qorishiq/jarangsiz
    'sh': ['š'],      # til oldi/sirg‘aluvchi/jarangsiz
    'g‘': ['γ', 'y']  # chuqur til orqa va oldi/sirg‘aluvchi/jarangli
}

# So‘zni analiz qilish funksiyasi
def analyze_word(word):
    result = ""
    i = 0
    while i < len(word):
        # Ikki harfli birikmalarni tekshirish (masalan, "ng", "ch")
        if i < len(word) - 1 and word[i:i+2] in sound_rules:
            key = word[i:i+2]
            result += sound_rules[key][0]  # Birinchi variantni olamiz
            i += 2
        # Bir harfli tovushlarni tekshirish
        elif word[i] in sound_rules:
            key = word[i]
            result += sound_rules[key][0]  # Birinchi variantni olamiz
            i += 1
        else:
            result += word[i]  # Agar qoidada bo‘lmasa, o‘zgarishsiz qoldiramiz
            i += 1
    return result

# Test qilish uchun foydalanuvchi kiritgan so‘zni qabul qilish
word = input("So‘zni kiriting: ")
transformed_word = analyze_word(word)
print(f"Kiritilgan so‘z: {word}")
print(f"O‘zgartirilgan shakl: {transformed_word}")

# Misollar bilan sinov
test_words = ["xatna", "taram", "qo‘y", "uyning", "chiroyli"]
for test in test_words:
    print(f"{test} -> {analyze_word(test)}")
```
